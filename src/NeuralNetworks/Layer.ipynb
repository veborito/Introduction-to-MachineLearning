```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# input vector
X = np.array([-1,0, 1])
n_features = len(X)
n_outputs = 1
# param matrix
B = np.random.normal(size = [n_outputs, n_features])
# intermediate output
z = np.matmul(B, X)
y_hat = sigmoid(z)
```

```python
plt.plot(np.arange(-10, 10), sigmoid(np.arange(-10, 10)))
plt.plot(np.arange(-10, 10), 1 - sigmoid(np.arange(-10, 10)))
plt.legend(["P(y = 1)", "P(y = -1)"])
plt.show()
```

```python
# Softmax optimization
z = np.array([1, 3, -1]) # correspond to 'scores' for the three differrent classes
# how do I get a probability vector that sums to one and is monotonic and everything is > 0 ?
y_hat = np.exp(z) / np.sum(np.exp(z))
```

```python
x = np.arange(-10, 10, 0.1)
def Huber(x, delta):
    r = np.abs(x) < delta
    return (0.5 * r * x**2 + (1 - r) * (delta * np.abs(x) - 0.5 *delta))
plt.plot(x, (x)**2)
plt.plot(x, Huber(x, 0.5))
```
